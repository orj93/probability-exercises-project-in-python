## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
```

```python
probabilidad_total = 0
for i in range(1000):
    dado1 = random.randint(1,6)
    dado2 = random.randint(1,6)
    suma = dado1 + dado2
    if suma % 2 == 0 or suma > 7:
        probabilidad_total += 1
        
resultado = probabilidad_total/1000
print(resultado)
```

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(1, 61):
    if i <= 10:
        ball_box[i] = "White"
        
    elif (i > 10) and (i < 31):
        ball_box[i] = "Red"
        
    else:
        ball_box[i] = "Green"

#print(ball_box)


probabilidad_total=0

for j in range(1000):
    seleccion_bolas = []    
    for i in range(1,6):
        seleccion_bolas.append(ball_box[random.randint(1,60)])
    
    #print(seleccion_bolas)
    conteo_rojas=0
    conteo_blancas=0
    conteo_verdes=0   
    
    

    for i in seleccion_bolas:
        if i == 'White':
            conteo_blancas += 1
        elif i == 'Red':
            conteo_rojas += 1
        elif i == 'Green':
            conteo_verdes += 1

    #print(conteo_blancas,conteo_rojas)

        if conteo_blancas==3 and conteo_rojas==2:
            probabilidad_total += 1
            #print(probabilidad_total)
    
print(probabilidad_total/1000)
```
